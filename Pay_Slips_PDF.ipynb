# **SCALE TO MULTIPLE PDFS**

```python
# Import required libraries
from pdfminer.high_level import extract_text
from google.colab import drive
import re
import pandas as pd
import os

# Mount Google Drive
drive.mount('/content/drive')

# Path to the folder containing PDF files on Google Drive
pdf_folder_path = "/content/drive/MyDrive/PaySlips"  # Replace 'path_to_your_pdf_folder' with the actual path to your folder

# Function to safely extract using regex
def safe_extract(pattern, text):
    match = re.search(pattern, text)
    return match.group(1) if match else "Not Found"

# Lists to store extracted details
employer_details_list = []
employee_details_list = []

# Loop through each file in the specified directory
for filename in os.listdir(pdf_folder_path):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder_path, filename)

        # Extract text from the PDF using pdfminer.six
        pdf_text = extract_text(pdf_path)

        # Extract and format specific sections
        employer_details = {
            "PDF Name": filename,
            "Name": safe_extract(r'EMPLOYEUR Nom : (.*?) Adresse', pdf_text),
            "Numéro APE": safe_extract(r'Numéro APE : (.*?) Numéro SIRET', pdf_text),
            "Numéro SIRET": safe_extract(r'Numéro SIRET : (.*?) SALARIE', pdf_text)
        }
        employer_details_list.append(employer_details)

        employee_details = {
            "PDF Name": filename,
            "Name": safe_extract(r'SALARIE Nom et Prénom : (.*?) Adresse', pdf_text),
            "Numéro SS": safe_extract(r'Numéro SS : (.*?) Date Entrée', pdf_text),
            "Date Entrée": safe_extract(r'Date Entrée : (.*?) Emploi', pdf_text),
            "Emploi": safe_extract(r'Emploi : (.*?) Salaire de base', pdf_text)
        }
        employee_details_list.append(employee_details)

# Convert lists to DataFrames
employer_df = pd.DataFrame(employer_details_list)
employee_df = pd.DataFrame(employee_details_list)

# Display the consolidated tables
print("Consolidated Employer Details:")
display(employer_df)

print("\nConsolidated Employee Details:")
display(employee_df)
```
